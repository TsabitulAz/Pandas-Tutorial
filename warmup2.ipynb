Adding or Removing Columns

```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
coffee = pd.read_csv('./data/coffee.csv')
bios = pd.read_csv('./data/bios.csv')
```

```python
coffee.head()
```

```python
coffee['Price'] = 4.99
coffee.head()
coffee.info()
```

```python
#Adding columns
#np.where() = membuat kolom baru dataframe berdasarkan perkondisian if else
coffee['New Price'] = np.where(coffee['Coffee Type'] == 'Espresso', 3.99, 5.99) 
#True = 3.99
#False = 5.99

checkingData = coffee[(coffee['Coffee Type'] == 'Espresso') &
                      (coffee['Units Sold'].isin([20, 25, 30])) &
                      (coffee['New Price'] == 3.99)][['Coffee Type', 'Units Sold', 'New Price']].reset_index(drop=True)

checkingData
```

```python
#Remove columns
coffee.drop(columns=['Price'], inplace=True) #inplace(True) -> menghapus secara permanen

#Bisa gunakan -> deletePrice = coffee.drop(columns=['Price'])
```

```python
coffee
```

```python
#Merapikan data sblm analisis atau ekspor
coffee = coffee[['Day', 'Coffee Type', 'Units Sold', 'New Price']]
```

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['New Price']
```

```python
coffee.rename(columns={'New Price': 'Price'}, inplace=True)
coffee
```

```python
#lambda function
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
bios.sample(5)
```

```python
#Membuat perkondisian suatu series menggunakan function def
def categorize_athlets(row):
    if row['height_cm'] < 170 and row['weight_kg'] < 70:
        return 'lightweight'
    elif row['height_cm'] < 180 and row['weight_kg'] < 80:
        return 'middleweight'
    else:
        return 'heavyweight'
    
bios['Category'] = bios.apply(categorize_athlets, axis=1) #axis=1(baris) untuk operasi yang membutuhkan data dari beberapa kolom dalam satu baris
bios.sample(5)
```

Merging and Concatenating Data

```python
nocs = pd.read_csv('./data/noc_regions.csv')
nocs.head()
```

```python
#pd.merge -> fungsi pandas untuk join dataframe 
#left_on = kolom dari dataframe pertama, right_on = kolom dari dataframe kedua
#how = jenis join yang digunakan (left, right, inner, outer)
bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
bios_new.head()
```

```python
bios_new.rename(columns={'region': 'born_country_full'}, inplace=True)
bios_new.sample(3)
```

```python
usa = bios[bios['born_country'] == 'USA'].copy()
fra = bios[bios['born_country'] == 'FRA'].copy()
usa.sample(5)
fra.sample(5)
```

```python
new_df = pd.concat([usa, fra])  # Menggabungkan dua DataFrame
new_df.sample(5)
```

Handling Null Values

```python
coffee.loc[[2,3], 'Units Sold'] = np.nan
coffee.loc[[2,3], 'Units Sold']  # Memeriksa nilai yang telah diubah menjadi NaN
```

```python
coffee['Units Sold'].fillna(coffee['Units Sold'].mean())  # Mengisi NaN dengan rata-rata
coffee.head()
```

```python
coffee.dropna(subset=['Units Sold'])  # Menghapus baris dengan NaN pada kolom 'Units Sold'
#Use inplace=True if you want to update the coffee df
```

```python
coffee[coffee['Units Sold'].notna()] #Memilih baris yang tidak memiliki NaN pada kolom 'Units Sold'
coffee[coffee['Units Sold'].isna()]  # Memilih baris yang memiliki NaN pada kolom 'Units Sold'
```

```python
#Menggunakan interpolasi untuk mengisi nilai NaN -> mengisi nilai NaN dengan nilai rata-rata dari nilai di sekitarnya
#coffee['Units Sold'] = coffee['Units Sold'].interpolate()  => Jika ingin menyimpan hasil interpolasi ke dalam kolom yang sama -> sama seperti inplace=True 
coffee['Units Sold'] = coffee['Units Sold'].interpolate()
coffee
```
